## Parte 1 — Módulo sencillo: `fisica_basica.py`

Contiene **funciones** para cinemática, energía y fuerza.  
No hay clases, solo funciones que reciben parámetros y retornan un valor.

````python
# Importar el modulo completo
import fisica_basica

from fisica_basica import G as aceleracion_gravitacional

# También se puede importar solo lo que necesitas:
from fisica_basica import posicion, energia_cinetica
````

````python
aceleracion_gravitacional
````

````python
# Ver las constantes definidas en el modulo
print("g =", fisica_basica.G, "m/s²")
print("c =", fisica_basica.C, "m/s")
````

### Cinemática: caída libre

Una pelota cae desde 45 m de altura. ¿Cuánto tarda en llegar al suelo? ¿Con qué velocidad?

````python
h = 45  # metros

t = fisica_basica.tiempo_caida(h)
v = fisica_basica.velocidad(v0=0, a=fisica_basica.G, t=t)

print(f"Tiempo de caída:     {t:.2f} s")
print(f"Velocidad al llegar: {v:.2f} m/s")
````

### Energía mecánica

Una pelota de 2 kg está a 10 m de altura y se mueve a 3 m/s.

````python
m, v, h = 2, 3, 10

K = fisica_basica.energia_cinetica(m, v)
U = fisica_basica.energia_potencial(m, h)
E = fisica_basica.energia_total(m, v, h)

print(f"Energía cinética:    {K:.2f} J")
print(f"Energía potencial:   {U:.2f} J")
print(f"Energía total:       {E:.2f} J")
````

### Movimiento con fuerza constante

Un objeto parte del reposo y se le aplica 20 N durante 5 s.

````python
m = 4    # kg
F = 20   # N
a = F / m

for t in [1, 2, 3, 4, 5]:
    x = fisica_basica.posicion(x0=0, v0=0, a=a, t=t)
    v = fisica_basica.velocidad(v0=0, a=a, t=t)
    print(f"t={t}s  →  x={x:.1f} m   v={v:.1f} m/s")
````

## Parte 2 — Módulo con POO: `fisica_poo.py`

Aquí los conceptos físicos se modelan como **objetos**.

```
CuerpoFisico          ← clase base
├── Particula         ← sin volumen, puede tener carga
└── ObjetoRigido      ← tiene volumen y densidad
    ├── Esfera
    └── Cubo
```

Cada subclase **hereda** los atributos y métodos de su clase padre y puede agregar los propios.

````python
from fisica_poo import Particula, Esfera, Cubo
````

### Partícula cargada en un campo eléctrico

````python
electron = Particula(nombre="electron", masa=9.11e-31, carga=-1.6e-19)
print(electron)

E_campo = 1000  # V/m
F = electron.fuerza_electrica(E_campo)
print(f"Fuerza eléctrica: {F:.3e} N")
````

### Objeto bajo una fuerza: herencia en acción

`Particula` hereda `aplicar_fuerza()` de `CuerpoFisico`.

````python
pelota = Particula(nombre="pelota", masa=0.5)
print("Antes:", pelota)

pelota.aplicar_fuerza(fuerza=10, tiempo=3)   # F=10 N durante 3 s
print("Después:", pelota)
print(f"Energía cinética: {pelota.energia_cinetica():.2f} J")
````

### Esfera de plomo

````python
esfera = Esfera(nombre="esfera de plomo", radio=0.05, densidad=11340)
print(esfera)
print(f"Área de superficie: {esfera.area_superficie():.4f} m²")
````

### Cubo de aluminio

````python
cubo = Cubo(nombre="cubo de aluminio", lado=0.1, densidad=2700)
print(cubo)
print(f"Área de superficie: {cubo.area_superficie():.4f} m²")
````

### Polimorfismo: misma función, distintos objetos

Como `Esfera` y `Cubo` heredan de la misma clase base,  
podemos tratarlos de forma uniforme.

````python
objetos = [
    Esfera("bola de acero", radio=0.03, densidad=7800),
    Cubo("dado de madera", lado=0.02, densidad=600),
    Esfera("canica de vidrio", radio=0.008, densidad=2500),
]

print(f"{'Objeto':<20} {'Masa (g)':>10} {'Volumen (cm³)':>15}")
print("-" * 48)
for obj in objetos:
    print(f"{obj.nombre:<20} {obj.masa*1000:>10.3f} {obj.volumen()*1e6:>15.3f}")
````
