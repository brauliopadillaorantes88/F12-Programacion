## ¿Qué es NumPy?

**NumPy** (Numerical Python) es la librería fundamental para cálculo numérico en Python.
Casi todo el ecosistema científico de Python (Pandas, Matplotlib, Scikit-learn, SciPy)
está construido sobre NumPy.

## ¿Por qué necesitamos NumPy si ya tenemos listas?

Excelente pregunta. Veamos el problema con un ejemplo concreto.

Imagina que tienes las temperaturas de 5 días y quieres convertirlas de Celsius a Fahrenheit
usando la fórmula `F = C * 9/5 + 32`.

**Con una lista de Python** necesitas un ciclo `for`:

````python
# Con lista de Python — necesitas un ciclo
celsius_lista = [0, 20, 37, 100, -10]
fahrenheit_lista = []

for c in celsius_lista:
    fahrenheit_lista.append(c * 9/5 + 32)

print('Fahrenheit (con lista):', fahrenheit_lista)
````

````python
import numpy as np

# Con NumPy — sin ciclos, más rápido y más legible
celsius_array = np.array([0, 20, 37, 100, -10])
fahrenheit_array = celsius_array * 9/5 + 32

print('Fahrenheit (con NumPy):', fahrenheit_array)
print(f'\nVersión de NumPy: {np.__version__}')
````

El resultado es el mismo, pero NumPy es:
- **Más legible**: la operación matemática se ve igual que la fórmula
- **Más rápido**: para millones de datos, NumPy puede ser 100x más rápido que un ciclo Python
- **Más compacto**: una línea en lugar de cuatro

Veamos la diferencia de velocidad con datos grandes:

````python
import time

N = 1_000_000  # un millón de datos
datos_lista = list(range(N))
datos_array = np.arange(N, dtype=float)

# Con lista y ciclo for
inicio = time.time()
resultado_lista = [x * 2 + 1 for x in datos_lista]
tiempo_lista = time.time() - inicio

# Con NumPy
inicio = time.time()
resultado_array = datos_array * 2 + 1
tiempo_numpy = time.time() - inicio

print(f'Lista (for loop): {tiempo_lista:.4f} segundos')
print(f'NumPy array:      {tiempo_numpy:.4f} segundos')
print(f'NumPy es {tiempo_lista/tiempo_numpy:.0f}x más rápido')
````

## 1. Creación de Arrays

Un **array** de NumPy es como una lista, pero:
- Todos los elementos tienen el **mismo tipo** (int, float, bool…)
- Puede tener múltiples dimensiones (1D, 2D, 3D…)
- Las operaciones matemáticas se aplican a **todos los elementos a la vez**

### 1.1 Desde una lista

````python
# Array 1D — como una fila de datos
notas = np.array([85, 72, 91, 60, 78, 95])

print('Array:', notas)
print('Tipo de datos (dtype):', notas.dtype)  # int64 — todos son enteros
print('Dimensiones (ndim):  ', notas.ndim)    # 1 — una sola dimensión
print('Forma (shape):       ', notas.shape)   # (6,) — 6 elementos
print('Total elementos:     ', notas.size)    # 6
````

````python
# Si mezclas int y float, NumPy convierte todo a float (el tipo más general)
mixto = np.array([1, 2.5, 3, 4.0])
print('Array mixto int/float:', mixto)
print('Dtype resultante:     ', mixto.dtype)  # float64
````

````python
# Array 2D — como una tabla (filas y columnas)
# Cada lista interna es una FILA
#
#           Mat  Fís  Quím
califs = np.array([
    [85, 90, 78],   # Estudiante 1
    [72, 88, 95],   # Estudiante 2
    [91, 76, 83],   # Estudiante 3
    [68, 94, 87]    # Estudiante 4
])

print('Tabla de calificaciones:')
print(califs)
print(f'\nShape: {califs.shape}  →  {califs.shape[0]} estudiantes, {califs.shape[1]} materias')
````

### 1.2 Funciones de creación

NumPy tiene funciones para crear arrays sin escribir cada valor manualmente.
Son muy útiles para inicializar arrays antes de llenarlos con un ciclo.

````python
# np.zeros: lleno de ceros
# Caso de uso: inicializar un array antes de llenarlo con un ciclo
acumulador = np.zeros(5)
print('Inicializado en ceros:', acumulador)

# Llenarlo con un ciclo
for i in range(5):
    acumulador[i] = (i + 1) ** 2   # cuadrados: 1, 4, 9, 16, 25

print('Después del ciclo:    ', acumulador)
````

````python
# np.arange(inicio, fin, paso) — como range() pero devuelve array
# IMPORTANTE: el valor 'fin' NO está incluido

dias = np.arange(1, 8)          # días de la semana
pares = np.arange(0, 20, 2)     # números pares
decimales = np.arange(0, 1.1, 0.25)  # con paso decimal

print('Días (1 al 7):     ', dias)
print('Pares hasta 18:    ', pares)
print('De 0 a 1 en 0.25:  ', decimales)
````

````python
# np.linspace(inicio, fin, n_puntos) — n puntos equiespaciados
# A diferencia de arange, AQUÍ SÍ se incluye el fin
# Caso de uso: crear el eje X para graficar una función

x = np.linspace(0, 10, 6)   # 6 puntos equiespaciados entre 0 y 10
print('linspace(0, 10, 6):', x)
print('Distancia entre puntos:', x[1] - x[0])

# Para graficar sen(x), necesitamos muchos puntos (curva suave)
x_fino = np.linspace(0, 2*np.pi, 1000)
print(f'\nPara graficar: {len(x_fino)} puntos entre 0 y 2π')
````

````python
# Diferencia visual entre arange y linspace
print('arange(0, 1, 0.333):  ', np.arange(0, 1, 1/3).round(3),
      '← puede no incluir el 1.0')
print('linspace(0, 1, 4):    ', np.linspace(0, 1, 4).round(3),
      '← SIEMPRE incluye 0 y 1')
````

### 1.3 Arrays aleatorios

Se usan para:
- **Simulaciones** (simular datos de un experimento)
- **Tests** (probar código con datos de prueba)
- **Estadística** (bootstrap, Monte Carlo)

La `seed` (semilla) hace que los números "aleatorios" sean reproducibles:
si tú y yo usamos la misma semilla, obtenemos los mismos números.

````python
rng = np.random.default_rng(seed=42)  # semilla fija → resultados reproducibles

# Distribución uniforme: cualquier valor entre a y b tiene la misma probabilidad
# Caso de uso: sorteos, asignar grupos aleatoriamente
uniforme = rng.random(6)   # 6 valores entre 0.0 y 1.0
print('Uniforme [0,1):     ', uniforme.round(3))

# Enteros aleatorios
# Caso de uso: simular tiradas de dado, muestras aleatorias
dados = rng.integers(1, 7, size=10)   # 10 tiradas de un dado de 6 caras
print('Tiradas de dado:    ', dados)

# Distribución normal: campana de Gauss (μ=media, σ=desviación estándar)
# Caso de uso: simular alturas, errores de medición, notas de examen
alturas = rng.normal(loc=170, scale=8, size=5)   # media=170cm, std=8cm
print('Alturas simuladas:  ', alturas.round(1))
````

````python
# Ejemplo: simular 1000 tiradas de un dado y contar cada resultado
rng2 = np.random.default_rng(0)
tiradas = rng2.integers(1, 7, size=1000)

print('Resultado de 1000 tiradas de dado:')
for cara in range(1, 7):
    conteo = np.sum(tiradas == cara)   # cuántas veces salió este número
    barra  = '█' * (conteo // 10)     # barra proporcional
    print(f'  Cara {cara}: {conteo:4d} veces  {barra}')

print(f'\nTotal tiradas: {len(tiradas)}')
print(f'Esperado por cara: ~{1000//6} veces')
````

## 2. Indexing y Slicing

"Indexing" = acceder a **un elemento** por su posición.  
"Slicing" = acceder a **un rango** de elementos.

**La regla de los índices negativos:** en Python (y NumPy), el índice `-1` es el último,
`-2` el penúltimo, etc. Es como contar desde el final.

### 2.1 Arrays 1D

````python
#              0     1     2     3     4     5     6
temps = np.array([15.2, 17.8, 20.1, 23.5, 25.0, 22.3, 18.6])
#             -7    -6    -5    -4    -3    -2    -1

print('Array:', temps)
print()
print('Primer elemento    [0]:', temps[0])    # 15.2
print('Último elemento   [-1]:', temps[-1])   # 18.6  ← muy útil
print('Penúltimo         [-2]:', temps[-2])   # 22.3

print()
# Slicing: [inicio:fin:paso]  — fin NO incluido
print('Primeros 3    [0:3]:', temps[0:3])
print('Del 2 al 4    [2:5]:', temps[2:5])
print('Últimos 3     [-3:]:', temps[-3:])
print('De 2 en 2    [::2]:', temps[::2])
print('Invertido    [::-1]:', temps[::-1])
````

````python
# Caso de uso: analizar una serie de temperaturas semanales
temp_semana = np.array([22, 24, 21, 19, 25, 27, 23])   # L M X J V S D
dias_nombre = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom']

print('=== Análisis de temperatura semanal ===')
print(f'Temperatura más alta: {temp_semana.max()}°C el {dias_nombre[np.argmax(temp_semana)]}')
print(f'Temperatura más baja: {temp_semana.min()}°C el {dias_nombre[np.argmin(temp_semana)]}')
print(f'Promedio:             {temp_semana.mean():.1f}°C')

# Días laborales (Lun-Vie) vs fin de semana
laborales  = temp_semana[:5]   # índices 0 al 4
finde      = temp_semana[5:]   # índices 5 y 6
print(f'\nPromedio días laborales:  {laborales.mean():.1f}°C')
print(f'Promedio fin de semana:   {finde.mean():.1f}°C')
````

### 2.2 Arrays 2D

Para arrays 2D usamos **dos índices separados por coma**: `array[fila, columna]`.
El símbolo `:` significa "todos".

````python
#           Día1  Día2  Día3  Día4  Día5
T = np.array([
    [22, 24, 21, 23, 25],   # Ciudad A (fila 0)
    [15, 16, 14, 17, 15],   # Ciudad B (fila 1)
    [30, 32, 29, 31, 33],   # Ciudad C (fila 2)
    [10, 11,  9, 12, 10]    # Ciudad D (fila 3)
])
ciudades = ['Ciudad A', 'Ciudad B', 'Ciudad C', 'Ciudad D']

print('Tabla de temperaturas:')
print(T)

print('\n--- Acceso por posición ---')
print('Temperatura Ciudad A día 1  [0, 0]:', T[0, 0])   # 22
print('Temperatura Ciudad C día 5  [2, 4]:', T[2, 4])   # 33

print('\n--- Slicing ---')
print('Toda la Ciudad B (fila 1)  [1, :]:', T[1, :])    # fila completa
print('Todas las ciudades, Día 3  [:, 2]:', T[:, 2])    # columna completa
print('Submatriz 2×2 (A,B × días 1,2):')
print(T[0:2, 0:2])
````

````python
# Caso de uso: calcular el promedio de cada ciudad usando axis
# axis=1 → operación a lo largo de las COLUMNAS (promedio de cada fila)
promedios = T.mean(axis=1)

print('Temperatura promedio por ciudad:')
for ciudad, prom in zip(ciudades, promedios):
    print(f'  {ciudad}: {prom:.1f}°C')

print('\nTemperatura promedio por día (axis=0):')
prom_dias = T.mean(axis=0)
for d, p in enumerate(prom_dias, start=1):
    print(f'  Día {d}: {p:.1f}°C')
````

### 2.3 Indexing booleano (filtrado)

Esta es una de las técnicas más poderosas de NumPy.
Funciona así:
1. Creas una **condición** → NumPy devuelve un array de `True`/`False`
2. Usas ese array como "máscara" para **seleccionar** solo los elementos `True`

Es como un ciclo `for` con un `if`, pero mucho más rápido.

````python
notas = np.array([45, 72, 88, 34, 91, 60, 55, 78, 95, 42])

# Equivalente al ciclo:
# for nota in notas:
#     if nota >= 60:
#         ...

mascara = notas >= 60
print('Notas:        ', notas)
print('Máscara ≥60:  ', mascara)   # True donde la condición se cumple
print('Aprobados:    ', notas[mascara])
print(f'Cantidad de aprobados: {mascara.sum()} de {len(notas)}')
````

````python
# Múltiples condiciones: & (y), | (o)
# IMPORTANTE: cada condición va entre paréntesis
print('Notas entre 70 y 89:', notas[(notas >= 70) & (notas <= 89)])
print('Reprobados o 100:   ', notas[(notas < 60) | (notas == 100)])

# np.where: versión vectorizada del if/else
# np.where(condición, valor_si_true, valor_si_false)
# Equivalente a: ["Aprobado" if n >= 60 else "Reprobado" for n in notas]
estado = np.where(notas >= 60, 'Aprobado', 'Reprobado')
print('\nEstado de cada nota:')
for nota, est in zip(notas, estado):
    print(f'  {nota:3d} → {est}')
````

````python
# Modificar valores con indexing booleano
datos = np.array([85, 102, 76, 115, 90, 88, 130, 70])
print('Datos originales:  ', datos)

# Limitar valores al rango [0, 100]: equivale a:
# for i in range(len(datos)):
#     if datos[i] > 100: datos[i] = 100
datos_copia = datos.copy()
datos_copia[datos_copia > 100] = 100
print('Datos limitados:   ', datos_copia)

# np.clip hace lo mismo de forma más limpia
print('Con np.clip:       ', np.clip(datos, 0, 100))
````

## 3. Operaciones Matemáticas

### 3.1 Operaciones elemento a elemento

Cuando aplicas `+`, `-`, `*`, `/` entre dos arrays del mismo tamaño,
NumPy lo hace **elemento a elemento**. Esto reemplaza ciclos `for` comunes.


````python
precios    = np.array([100.0, 250.0, 80.0, 320.0, 60.0])
cantidades = np.array([3,     2,     5,    1,     10   ])

# Calcular el total de cada producto
# Equivalente a: [p * q for p, q in zip(precios, cantidades)]
totales = precios * cantidades

print('Precios:    ', precios)
print('Cantidades: ', cantidades)
print('Totales:    ', totales)
print(f'\nTotal general: Q{totales.sum():,.2f}')
print(f'Producto más caro total: Q{totales.max():.2f}')
````

````python
# Operación con escalar: se aplica a TODOS los elementos
# Caso de uso: aplicar un descuento del 15% a todos los precios
precios_con_descuento = precios * (1 - 0.15)
ahorro = precios - precios_con_descuento

print('Precio original:    ', precios)
print('Precio con -15%:    ', precios_con_descuento)
print('Ahorro:             ', ahorro.round(2))
````

### 3.2 Broadcasting — operaciones entre arrays de distinto tamaño

NumPy puede operar arrays de formas distintas siguiendo reglas automáticas.
La regla básica: el array más pequeño se "expande" para coincidir con el grande.

````python
# Tabla de ventas: 3 vendedores × 4 meses
ventas = np.array([
    [1200, 1500, 1100, 1800],   # Vendedor A
    [ 900, 1200,  950, 1100],   # Vendedor B
    [1500, 1800, 1400, 2000]    # Vendedor C
])

# Meta mensual distinta cada mes (vector de 4 elementos)
meta = np.array([1000, 1300, 1000, 1500])

# ¿Cuánto superó cada vendedor la meta cada mes?
# Broadcasting: meta (4,) se aplica a cada fila de ventas (3,4)
diferencia = ventas - meta

print('Ventas:')
print(ventas)
print('\nDiferencia respecto a la meta (+ = superó, - = no llegó):')
print(diferencia)

print('\n¿Cada vendedor cumplió la meta cada mes?')
print(ventas >= meta)
````

### 3.3 Funciones matemáticas (ufuncs)

NumPy tiene versiones vectorizadas de funciones matemáticas que actúan sobre
todos los elementos de un array a la vez.

````python
# Caso de uso: calcular la hipotenusa de varios triángulos a la vez
# hipotenusa = sqrt(a² + b²)
a = np.array([3, 5, 8, 12, 7])
b = np.array([4, 12, 6, 5, 24])

hipotenusa = np.sqrt(a**2 + b**2)
print('Cateto a:', a)
print('Cateto b:', b)
print('Hipotenusa:', hipotenusa)
````

````python
# Función seno para graficar una onda
# Los ángulos en NumPy siempre van en RADIANES (no grados)
# Para convertir: rad = grados * π/180

angulos_grados = np.array([0, 30, 45, 60, 90, 120, 180, 270, 360])
angulos_rad    = angulos_grados * np.pi / 180

print('Grados  Radianes   sin      cos')
print('-' * 40)
for g, r in zip(angulos_grados, angulos_rad):
    print(f'{g:6d}   {r:7.4f}   {np.sin(r):6.3f}   {np.cos(r):6.3f}')
````

## 4. Estadísticas

NumPy tiene funciones estadísticas muy eficientes. El parámetro `axis` controla
si la operación va **por columnas** (`axis=0`) o **por filas** (`axis=1`).

````python
rng = np.random.default_rng(7)
notas = rng.integers(50, 101, size=30)   # 30 notas entre 50 y 100

print('Notas:')
print(notas)
print()

# Estadísticos básicos
stats = [
    ('Cantidad',         len(notas)),
    ('Media',            np.mean(notas)),
    ('Mediana',          np.median(notas)),
    ('Desv. estándar',   np.std(notas)),
    ('Mínimo',           np.min(notas)),
    ('Máximo',           np.max(notas)),
    ('Q1 (25%)',         np.percentile(notas, 25)),
    ('Q3 (75%)',         np.percentile(notas, 75)),
]

for nombre, valor in stats:
    print(f'  {nombre:<20}: {valor:.2f}' if isinstance(valor, float) else
          f'  {nombre:<20}: {valor}')
````

````python
# Estadísticas con axis en una matriz
# Notas de 4 estudiantes en 3 exámenes
#              Examen1  Examen2  Examen3
notas_mat = np.array([
    [75, 82, 90],   # Est. 1
    [60, 70, 65],   # Est. 2
    [95, 88, 92],   # Est. 3
    [80, 76, 84]    # Est. 4
])

# axis=1: promedio de cada estudiante (a lo largo de los exámenes)
prom_estud = notas_mat.mean(axis=1)
# axis=0: promedio de cada examen (a lo largo de los estudiantes)
prom_exam  = notas_mat.mean(axis=0)

print('Promedio por estudiante:')
for i, p in enumerate(prom_estud, 1):
    estado = 'Aprobado' if p >= 70 else 'Reprobado'
    print(f'  Est. {i}: {p:.1f} → {estado}')

print('\nPromedio por examen:')
for i, p in enumerate(prom_exam, 1):
    print(f'  Examen {i}: {p:.1f}')
````

## 5. Reshape — Cambiar la forma

`reshape` reorganiza los elementos en una nueva forma **sin copiarlos**.
El número total de elementos debe ser el mismo.

**Casos de uso:**
- Transformar datos 1D en una tabla 2D
- Preparar datos para modelos de Machine Learning (que esperan formas específicas)

````python
# Imagina que recibes datos de temperatura de 12 meses de forma lineal
temp_anual = np.array([22,19,21,24,27,30,31,30,28,25,22,19])
meses = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']

print('Datos originales (1D):', temp_anual)

# Reorganizar en 4 trimestres × 3 meses
trimestres = temp_anual.reshape(4, 3)
print('\nPor trimestres (4×3):')
for i, trimestre in enumerate(trimestres, 1):
    print(f'  T{i}: {trimestre}  → promedio: {trimestre.mean():.1f}°C')
````

````python
# -1 le dice a NumPy: "calcula tú esta dimensión"
v = np.arange(1, 13)
print('Original (12 elementos):', v)
print('reshape(3, -1):')
print(v.reshape(3, -1))   # NumPy calcula: 12/3 = 4 columnas
print('reshape(-1, 6):')
print(v.reshape(-1, 6))   # NumPy calcula: 12/6 = 2 filas
````

## 6. Álgebra Lineal

El módulo `np.linalg` implementa operaciones de álgebra lineal.

**Atención:** `A * B` es multiplicación **elemento a elemento** (no es el producto matricial).  
Para el producto matricial usa `A @ B` o `np.matmul(A, B)`.

````python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print('A:')
print(A)
print('B:')
print(B)
print()
print('A * B  (elemento a elemento — NO es producto matricial):')
print(A * B)
print()
print('A @ B  (producto MATRICIAL — esto es álgebra lineal):')
print(A @ B)
````

````python
# Caso de uso: resolver un sistema de ecuaciones
#
# Un estudiante compra 2 lápices y 3 cuadernos y paga Q17.
# Otro estudiante compra 4 lápices y 1 cuaderno y paga Q11.
# ¿Cuánto cuesta cada lápiz y cada cuaderno?
#
# Sistema de ecuaciones:
#   2x + 3y = 17
#   4x + 1y = 11

A = np.array([[2, 3],   # coeficientes
              [4, 1]])
b = np.array([17, 11])  # resultados

solucion = np.linalg.solve(A, b)
print(f'Precio del lápiz:   Q{solucion[0]:.2f}')
print(f'Precio del cuaderno: Q{solucion[1]:.2f}')

# Verificar: A @ solucion debe dar b
print(f'\nVerificación: A @ x = {A @ solucion}  (debe ser {b})')
````

## 7. Funciones Útiles con Ciclos

A veces necesitamos combinar NumPy con ciclos `for` para lógica más compleja.

````python
# np.argsort devuelve los ÍNDICES que ordenarían el array
# Caso de uso: encontrar el ranking de estudiantes por nota

nombres = np.array(['Ana', 'Luis', 'María', 'Carlos', 'Sofía', 'Pedro'])
notas   = np.array([78, 91, 65, 88, 95, 72])

# argsort en orden descendente (nota más alta = puesto 1)
ranking_indices = np.argsort(notas)[::-1]  # ::-1 invierte

print('=== Ranking de notas ===')
for puesto, idx in enumerate(ranking_indices, start=1):
    print(f'  #{puesto}  {nombres[idx]:<8}  {notas[idx]} puntos')
````

````python
# np.unique: valores únicos y su frecuencia
# Caso de uso: contar cuántos estudiantes hay por zona
zonas = np.array([1, 3, 1, 2, 5, 3, 1, 2, 1, 7, 3, 5, 2, 1, 3])

valores_unicos, conteos = np.unique(zonas, return_counts=True)

print('Distribución de estudiantes por zona:')
for zona, conteo in zip(valores_unicos, conteos):
    barra = '▓' * conteo
    print(f'  Zona {zona}: {conteo} estudiantes  {barra}')
````

## 8. Normalización y Estandarización

Dos técnicas fundamentales antes de hacer cálculos o Machine Learning.

**Normalización Min-Max:** escala los datos al rango [0, 1]
$$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

**Estandarización Z-score:** media=0, desviación estándar=1
$$z = \frac{x - \mu}{\sigma}$$

````python
# Queremos comparar alturas (cm) y pesos (kg) de estudiantes
# No podemos compararlos directamente porque tienen escalas distintas
alturas = np.array([165, 170, 175, 160, 180, 155, 172])
pesos   = np.array([ 60,  70,  80,  55,  85,  50,  75])

# Normalización Min-Max
def normalizar(x):
    return (x - x.min()) / (x.max() - x.min())

# Estandarización Z-score
def estandarizar(x):
    return (x - x.mean()) / x.std()

alt_norm = normalizar(alturas)
pes_norm = normalizar(pesos)

print('Datos originales y normalizados:')
print(f'{"Altura":>8} {"Peso":>6} {"Alt_norm":>10} {"Pes_norm":>10}')
print('-' * 40)
for a, p, an, pn in zip(alturas, pesos, alt_norm, pes_norm):
    print(f'{a:8d} {p:6d} {an:10.3f} {pn:10.3f}')
````

## 🏋️ Ejercicio 2: Simulación de Calificaciones

Simula las calificaciones de un curso con 40 estudiantes en 4 exámenes.

**Tareas:**
1. Genera los datos con `rng.integers(50, 101, size=(40, 4))`
2. Calcula el promedio de cada estudiante
3. Asigna letra: A (≥90), B (≥80), C (≥70), D (≥60), F (<60)
4. Imprime cuántos estudiantes hay con cada letra
5. Imprime el promedio del examen más difícil (el de menor promedio)


````python
rng = np.random.default_rng(2025)

# 1. Generar datos
examenes = rng.integers(50, 101, size=(40, 4)).astype(float)

# 2. Promedio por estudiante (axis=1: a lo largo de los exámenes)
promedios = examenes.mean(axis=1)

# 3. Asignar letra con ciclo for y condicionales
def asignar_letra(prom):
    if prom >= 90:
        return 'A'
    elif prom >= 80:
        return 'B'
    elif prom >= 70:
        return 'C'
    elif prom >= 60:
        return 'D'
    else:
        return 'F'

letras = np.array([asignar_letra(p) for p in promedios])

# 4. Conteo por letra
print('=== Distribución de calificaciones ===')
for letra in ['A', 'B', 'C', 'D', 'F']:
    conteo  = np.sum(letras == letra)
    pct     = conteo / len(letras) * 100
    barra   = '█' * conteo
    print(f'  {letra}: {conteo:2d} estudiantes ({pct:4.1f}%)  {barra}')

# 5. Examen más difícil
prom_examenes = examenes.mean(axis=0)  # promedio de cada examen
idx_dificil   = np.argmin(prom_examenes)
print(f'\nPromedios por examen: {prom_examenes.round(1)}')
print(f'Examen más difícil: Examen {idx_dificil+1} (promedio: {prom_examenes[idx_dificil]:.1f})')
````
