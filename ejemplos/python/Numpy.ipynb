# NumPy — Computación Numérica en Python

## ¿Qué es NumPy?

**NumPy** (Numerical Python) es la librería fundamental para computación científica en Python. Proporciona:

- Un objeto de arreglo (array) multidimensional de alto rendimiento
- Herramientas para trabajar con estos arreglos
- Funciones matemáticas, lógicas, de ordenamiento, estadísticas, y más

## ¿Para qué sirve?

| Tarea | Sin NumPy | Con NumPy |
|---|---|---|
| Sumar dos listas de 1 millón de elementos | ~0.5 segundos | ~0.005 segundos |
| Multiplicar matrices | Código complejo con loops | Una línea |
| Operaciones matemáticas | Loop por cada elemento | Operaciones vectorizadas |

NumPy es la base de casi todas las librerías de ciencia de datos: Pandas, Matplotlib, Scikit-learn, etc.

## Nivel 1 — Conceptos Básicos: Arrays

### ¿Qué diferencia hay entre una lista de Python y un array de NumPy?

````python
# Lista de Python
lista = [1, 2, 3, 4, 5]
print("Lista Python:", lista)
print("Tipo:", type(lista))

print()

# Array de NumPy
arreglo = np.array([1, 2, 3, 4, 5])
print("Array NumPy:", arreglo)
print("Tipo:", type(arreglo))
````

````python
# La diferencia clave: operaciones elemento a elemento
lista_python = [1, 2, 3]

# Con listas, el operador * repite la lista
print("Lista * 3:", lista_python * 3)  # [1, 2, 3, 1, 2, 3, 1, 2, 3]

arreglo_np = np.array([1, 2, 3])

# Con NumPy, * multiplica cada elemento
print("Array * 3:", arreglo_np * 3)   # [3, 6, 9]
````

### ¿Por qué NumPy es tan rápido? — Comparación de rendimiento

NumPy implementa las operaciones sobre arrays en **código C precompilado**. La misma operación que requiere un ciclo `for` en Python se ejecuta internamente sin ciclos explícitos. Esto se conoce como *código vectorizado* (Stewart & Mommert, §4.1).

````python
import time

N = 10_000_000   # 10 millones de elementos

# ── Enfoque con lista Python ──────────────────────────────────────────────────
a_lista = list(range(N))
t0 = time.time()
b_lista = []
for i in range(len(a_lista)):
    b_lista.append(a_lista[i] + 1)
t_lista = time.time() - t0

# ── Enfoque con array NumPy ───────────────────────────────────────────────────
a_np = np.array(a_lista)
t0 = time.time()
b_np = a_np + 1          # ← una sola línea, sin ciclo explícito
t_numpy = time.time() - t0

print(f"Lista Python : {t_lista:.3f} segundos")
print(f"Array NumPy  : {t_numpy:.3f} segundos")
print(f"NumPy es {t_lista / t_numpy:.0f}× más rápido")
````

### Tipos de datos (`dtype`) y atributos del array

Todos los elementos de un array NumPy tienen el **mismo tipo de dato** (`dtype`). NumPy infiere el tipo al crear el array, pero puedes especificarlo. Tipos comunes: `int64`, `float64`, `complex128`, `bool`. Puedes convertir con `.astype()` (Stewart & Mommert, §4.1.1).

````python
x = np.array([1, 2, 3.268999])         # Python convierte todo a float64
print("dtype inferido:", x.dtype)

# Especificar dtype manualmente
x_int  = np.array([1.5566666666, 2, 3], dtype=int)
x_cpx  = np.array([1, 2, 3], dtype=complex)
print("dtype=int    :", x_int,  "→", x_int.dtype)
print("dtype=complex:", x_cpx, "→", x_cpx.dtype)

# Convertir dtype con astype()
x_float = x_int.astype(float)
print("\nx_int.astype(float):", x_float, "→", x_float.dtype)

# Atributos importantes de un array
m = np.array([[1, 2, 3], [4, 5, 6]])
print("\nArray m:")
print(m)
print("  ndim :", m.ndim)    # número de dimensiones
print("  shape:", m.shape)   # (filas, columnas)
print("  size :", m.size)    # total de elementos
print("  dtype:", m.dtype)   # tipo de dato
````

### Broadcasting — reglas para operar arrays de distinta forma

Cuando aplicas una operación entre un array y un escalar (o dos arrays de forma distinta), NumPy ajusta los tamaños automáticamente siguiendo **dos reglas** (Stewart & Mommert, §4.1.4):

1. **Regla 1** — Si los arrays tienen distinto número de dimensiones, la forma del más pequeño se *extiende* a la izquierda con "1" hasta igualar.
2. **Regla 2** — Un eje de tamaño 1 se comporta como si tuviera el tamaño del eje más grande en esa dimensión (se "replica").

```
array   →  [1, 2, 3, 4, 5]   forma: (5,)
escalar →   10               forma: () → (1,) → (5,)   [reglas 1 y 2]
resultado→ [11, 12, 13, 14, 15]
```

Si dos ejes no son 1 y son distintos entre sí → **ValueError** (no se puede hacer broadcast).

````python
v = np.array([1, 2, 3, 4, 5])

print("Array original:", v)
print("v + 10     :", v + 10)    # suma 10 a cada elemento
print("v * 3      :", v * 3)     # multiplica cada elemento por 3
print("v ** 2     :", v ** 2)    # eleva al cuadrado cada elemento
print("v / 2      :", v / 2)     # divide cada elemento entre 2
print("v - 1      :", v - 1)

# También funciona con matrices
m = np.array([[1, 2, 3],
              [4, 5, 6]])
print("\nMatriz + 100:")
print(m + 100)
print("\nMatriz * 2:")
print(m * 2)
````

### Funciones matemáticas (ufuncs)

NumPy tiene versiones de las funciones matemáticas de `math` que funcionan directamente sobre **arrays completos** sin necesidad de un ciclo `for`. Estas se llaman *ufuncs* (universal functions).

````python
x = np.array([0, 1, 2, 3, 4, 5])

print("x          :", x)
print("np.sqrt(x) :", np.sqrt(x))      # raíz cuadrada de cada elemento
print("np.exp(x)  :", np.exp(x))       # e^x
print("np.log(x+1):", np.log(x + 1))  # logaritmo natural (evitamos log(0))
print("np.abs(x-3):", np.abs(x - 3))  # valor absoluto
print("np.sign(x-3):", np.sign(x - 3)) # -1, 0 o 1 según el signo
````

````python
# Funciones trigonométricas (el argumento está en radianes)
angulos = np.array([0, np.pi/6, np.pi/4, np.pi/3, np.pi/2])

print("Ángulos (rad):", angulos.round(4))
print("sin          :", np.sin(angulos).round(4))
print("cos          :", np.cos(angulos).round(4))

# Caso práctico: calcular la componente vertical de lanzamientos
# con distintos ángulos y velocidad inicial v0 = 10 m/s
v0 = 10
v_vertical = v0 * np.sin(angulos)
print("\nVelocidad vertical (v0=10):", v_vertical.round(3))
````

## Nivel 3b — Valores especiales y operadores lógicos

### Valores especiales: `nan` e `inf`

Algunas operaciones producen resultados especiales:
- `np.nan` (*Not a Number*): resultado indefinido, como `log(-1)` o `0/0`
- `np.inf` (*infinito*): resultado de dividir un número entre cero

NumPy tiene versiones de las funciones estadísticas que **ignoran** los `nan`:
`np.nanmean`, `np.nansum`, `np.nanmax`, `np.nanmin`

````python
print("np.pi  =", np.pi)
print("np.e   =", np.e)
print("np.inf =", np.inf)
print("np.nan =", np.nan)

# nan aparece en operaciones indefinidas
x = np.array([-4.0, -1.0, 0.0, 1.0, 4.0])
print("\nsqrt de", x, "→", np.sqrt(x))   # sqrt de negativos → nan

# Funciones que ignoran nan
notas = np.array([85.0, np.nan, 72.0, np.nan, 91.0])
print("\nNotas con nan:", notas)
print("Media normal   :", np.mean(notas))       # ← devuelve nan
print("Media sin nan  :", np.nanmean(notas))    # ← ignora nan
print("Suma sin nan   :", np.nansum(notas))
````

### Operadores lógicos sobre arrays

Puedes comparar un array con un valor y obtienes un array de booleanos (`True`/`False`).
Luego ese array booleano puede usarse como filtro para seleccionar elementos.

````python
temps = np.array([18.5, 22.0, 35.1, 15.3, 28.7, 12.0, 30.0, 19])

# Comparación → array de True/False
print("Temperaturas:", temps)
print("temps > 25  :", temps > 25)

# Usar el array booleano para filtrar
calurosos = temps[temps > 25]
print("\nDías calurosos (> 25°C):", calurosos)

# Condición compuesta: & (y) , | (o)
moderados = temps[(temps >= 20) & (temps <= 30)]
print("Días moderados (20-30°C):", moderados)

# Contar cuántos cumplen la condición
print("\n¿Cuántos días > 25?:", np.sum(temps > 25))
print("¿Algún día < 10?  :", np.any(temps < 10))
print("¿Todos > 10?      :", np.all(temps > 10))
````

#### `np.where` — asignar valores según condición

`np.where(condición, valor_si_True, valor_si_False)` crea un array nuevo aplicando una regla a cada elemento. Es el equivalente vectorizado del operador ternario.

````python
notas = np.array([55, 72, 88, 45, 91, 63, 78])

# Etiquetar cada nota como "Aprobado" o "Reprobado"
estados = np.where(notas >= 61, "Aprobado", "Reprobado")
print(estados)

for n, e in zip(notas, estados):
    print(f"  Nota {n:3d} → {e}")

# También funciona con números
# Reemplazar notas menores a 61 con 0 (para penalización)
penalizadas = np.where(notas >= 61, notas, 0)
print("\nNotas penalizadas:", penalizadas)
````

### Copias vs Vistas (*views*) — diferencia crítica con listas

En NumPy existen **tres comportamientos** distintos al "asignar" un array:

| Operación | Resultado | Modifica original |
|---|---|---|
| `b = a` | **alias** — misma memoria | Sí |
| `b = a[1:4]` | **vista** (*view*) — misma memoria | **Sí** ⚠️ |
| `b = a.copy()` | **copia** — memoria nueva | No |

> **¡Diferencia clave con listas!** En Python, `l[1:4]` es siempre una copia. En NumPy, `a[1:4]` es una **vista** del array original — modificarla cambia el original. Esto es fuente frecuente de errores (Stewart & Mommert, §4.1.2).

````python
original = np.array([1, 2, 3, 4, 5])

# ⚠ ALIAS — dos nombres, un solo array
alias = original
alias[0] = 999
print("Alias modifica 'original':", original)

# ⚠ VISTA (slice) — ¡diferente a listas Python!
original2 = np.array([1, 2, 3, 4, 5])
vista = original2[1:4]      # NO es copia — es vista
print(vista)
vista[0] = 777
print("Slice (vista) modifica 'original2':", original2)

# ✅ COPIA independiente — usa .copy()
original3 = np.array([1, 2, 3, 4, 5])
copia = original3.copy()
copia[0] = 999
print("Copia NO modifica 'original3':", original3)
copia
````

### Unir arrays — concatenate, hstack, vstack, stack

NumPy ofrece varias formas de unir arrays (Stewart & Mommert, §4.1.5):

- `np.concatenate()`: función **general** — une a lo largo de un eje existente.
- `np.hstack()`: atajo para unir **horizontalmente** (eje 1 en 2D, eje 0 en 1D).
- `np.vstack()`: atajo para unir **verticalmente** (eje 0).
- `np.stack()`: apila creando un **nuevo eje**.

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Función general: np.concatenate
print("concatenate:", np.concatenate((a, b)))
print("hstack     :", np.hstack((a, b)))   # equivalente para 1D

# Para matrices
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])

print("\nvstack (apilar filas, axis=0):")
print(np.vstack((m1, m2)))

print("\nhstack (unir columnas, axis=1):")
print(np.hstack((m1, m2)))

print("\nstack — crea NUEVO eje (shape resultante: 2×2×2):")
apilado = np.stack((m1, m2), axis=0)
print(apilado)
print("Shape:", apilado.shape)
````

### Ordenar y buscar — `np.sort` y `np.argsort`

- `np.sort(x)` — retorna una **copia ordenada** (no modifica `x`).
- `np.argsort(x)` — retorna los **índices** que ordenarían `x`. Muy útil para ordenar arrays relacionados (e.g., ordenar nombres por nota) (Stewart & Mommert, §4.2.5).

````python
notas = np.array([78, 45, 92, 61, 88, 34, 73])
print("Notas originales:", notas)

# np.sort — retorna una copia ordenada (no modifica el original)
print("Ordenadas (asc) :", np.sort(notas))
print("Ordenadas (desc):", np.sort(notas)[::-1])

# np.argsort — retorna los ÍNDICES que ordenarían el array
indices = np.argsort(notas)
print("\nÍndices (argsort):", indices)
print("Aplicando índices :", notas[indices])   # equivale a sort

# Caso práctico: obtener ranking de estudiantes
nombres = np.array(["Ana", "Carlos", "Diana", "Eduardo", "Fátima", "Gabriel", "Helena"])
ranking = np.argsort(notas)[::-1]   # mayor a menor

print(ranking)
print("\nRanking:")
for pos, i in enumerate(ranking, 1):
    print(f"  {pos}. {nombres[i]}: {notas[i]}")
````

### Suma acumulativa — `np.cumsum`

`np.cumsum` calcula la suma acumulada: cada elemento del resultado es la suma de todos los anteriores más el actual. Útil para calcular totales progresivos (gastos, puntajes acumulados, etc.).

````python
puntos_por_nivel = np.array([100, 250, 180, 320, 210])

print("Puntos por nivel   :", puntos_por_nivel)
print("Puntos acumulados  :", np.cumsum(puntos_por_nivel))

# Ejemplo: gastos diarios y total acumulado
gastos = np.array([50, 120, 30, 85, 200, 45])
print("\nGastos diarios     :", gastos)
print("Total acumulado    :", np.cumsum(gastos))
print("Total de la semana :", np.sum(gastos))
````

### Estadísticas avanzadas — histograma, media ponderada, correlación

- `np.histogram(x, bins)` — distribución de frecuencias (retorna conteo y bordes).
- `np.average(x, weights=w)` — **media ponderada** (cada dato tiene un peso diferente).
- `np.corrcoef(x, y)` — **coeficiente de correlación** de Pearson entre dos arrays.
- **API moderna de números aleatorios**: usar `default_rng(semilla)` en lugar de `np.random.seed()` (Stewart & Mommert, §4.4).

````python
from numpy.random import default_rng
rng = default_rng(42)   # semilla para reproducibilidad

# Generar datos con distribución normal (media=70, desv=10)
calificaciones = rng.normal(loc=70, scale=10, size=200).clip(0, 100)

# np.histogram — distribución de frecuencias
conteo, bordes = np.histogram(calificaciones, bins=10)
print("Conteo por intervalo:", conteo)
print("Bordes de intervalos:", bordes.round(1))

# Media ponderada: notas con distintos pesos
mediciones = np.array([3.1, 2.7, 2.5, 3.1, 2.9])
incertidumbres = np.array([0.05, 0.15, 0.60, 0.10, 0.05])
print("\nMedia simple    :", np.mean(mediciones).round(4))
print("Media ponderada :", np.average(mediciones, weights=1/incertidumbres**2).round(4))
# Pesos = 1/σ² → mayor peso a mediciones más precisas

# Correlación entre dos arrays
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.1, 3.9, 6.2, 7.8, 10.1])
print("\nCoeficiente de correlación Pearson:")
print(np.corrcoef(x, y).round(4))
````

### Multiplicación de matrices y operaciones esenciales

| Operación | Sintaxis | Descripción |
|---|---|---|
| Elemento a elemento | `A * B` | Hadamard product |
| Matricial | `A @ B` o `np.matmul(A, B)` | Producto matricial |
| Transpuesta | `A.T` o `np.transpose(A)` | Intercambia filas↔columnas |
| Identidad | `np.identity(n)` | Matriz identidad n×n |
| Aplanar | `A.ravel()` o `np.ravel(A)` | Convierte a vector 1D |

````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("A * B (elemento a elemento):")
print(A * B)

print("\nA @ B (multiplicación matricial):")
print(A @ B)

print("\nA.T (transpuesta):")
print(A.T)

print("\nnp.identity(3) — matriz identidad:")
print(np.identity(3))

# Aplanar una matriz a vector 1D
print("\nA.ravel() — aplanar a 1D:", A.ravel())
````

## Nivel 3d — Álgebra Lineal (`numpy.linalg`)

NumPy incluye el módulo `numpy.linalg` (importado como `la`) para operaciones matriciales avanzadas (Stewart & Mommert, §4.7):

| Función | Descripción |
|---|---|
| `la.det(A)` | Determinante de A |
| `la.inv(A)` | Inversa de A (si det ≠ 0) |
| `la.norm(v)` | Norma euclidiana de un vector |
| `la.solve(A, b)` | Resuelve el sistema **Ax = b** |
| `la.eig(A)` | Eigenvalores y eigenvectores de A |

> `la.solve(A, b)` es numéricamente más estable y eficiente que calcular `la.inv(A) @ b`.

````python
import numpy.linalg as la

A = np.array([[4, 2, 0],
              [9, 3, 7],
              [1, 2, 1]], dtype=float)

print("Matriz A:")
print(A)

# Determinante
print("\nDeterminante:", la.det(A).round(4))

# Inversa (si det ≠ 0)
A_inv = la.inv(A)
print("\nInversa A⁻¹:")
print(A_inv.round(4))

# Verificación: A @ A⁻¹ debe ser la identidad
print("\nA @ A⁻¹ (≈ identidad):")
print((A @ A_inv).round(10))

# Norma de un vector
v = np.array([3.0, 4.0])
print("\nNorma del vector [3, 4]:", la.norm(v))   # = 5.0
````

````python
import numpy.linalg as la

# ─── Resolución de sistemas de ecuaciones: Ax = b ───────────────────────────
# Ejemplo: sistema de 3 ecuaciones
#   3x + 2y +  z = 5
#   5x + 5y + 5z = 5
#    x + 4y + 6z = -3
A = np.array([[3, 2, 1],
              [5, 5, 5],
              [1, 4, 6]], dtype=float)
b = np.array([5, 5, -3], dtype=float)

x = la.solve(A, b)
print("Solución x:", x.round(4))
print("Verificación Ax - b ≈ 0:", (A @ x - b).round(10))

# ─── Eigenvalores y eigenvectores ───────────────────────────────────────────
print("\n--- Eigenvalores ---")
M = np.array([[-2, -4,  2],
              [-2,  1,  2],
              [ 4,  2,  5]], dtype=float)
eigenvalores, eigenvectores = la.eig(M)
print("Eigenvalores:", eigenvalores.round(3))
print("Primer eigenvector:", eigenvectores[:, 0].round(3))
````

## Nivel 3e — Polinomios

NumPy permite representar polinomios por sus coeficientes y ofrece funciones para ajustar, evaluar y manipularlos (Stewart & Mommert, §4.6):

- `np.polyfit(x, y, deg)` — ajusta un polinomio de grado `deg` a datos *(x, y)* por mínimos cuadrados. Retorna los coeficientes `[a_n, ..., a_1, a_0]`.
- `np.polyval(coefs, x)` — evalúa el polinomio con coeficientes `coefs` en los puntos `x`.
- `np.poly(roots)` — construye coeficientes desde las raíces.
- `np.roots(coefs)` — calcula las raíces a partir de los coeficientes.

````python
# Datos: velocidad inicial vs distancia máxima (proyectil simplificado)
# Queremos ajustar un polinomio de grado 2: d = a*v^2 + b*v + c
v = np.array([10, 20, 30, 40, 50], dtype=float)   # velocidad (m/s)
d = np.array([5.1, 20.4, 45.9, 81.6, 127.5])      # distancia (m)

# np.polyfit: ajuste de mínimos cuadrados
coefs = np.polyfit(v, d, deg=2)
print("Coeficientes [a, b, c]:", coefs.round(4))
print(f"Modelo: d = {coefs[0]:.4f}·v² + {coefs[1]:.4f}·v + {coefs[2]:.4f}")

# np.polyval: evaluar el polinomio en nuevos valores
v_nuevo = np.array([25, 35, 45])
d_pred = np.polyval(coefs, v_nuevo)
print("\nPredicciones:")
for vi, di in zip(v_nuevo, d_pred):
    print(f"  v={vi} m/s → d≈{di:.2f} m")
````

````python
import matplotlib.pyplot as plt

# Curva para graficar
v_fit = np.linspace(0, 55, 100)
d_fit = np.polyval(coefs, v_fit)

# Gráfica
plt.figure()
plt.scatter(v, d, label="Datos reales")
plt.plot(v_fit, d_fit, label="Ajuste cuadrático")
plt.xlabel("Velocidad (m/s)")
plt.ylabel("Distancia (m)")
plt.title("Ajuste cuadrático: distancia vs velocidad")
plt.legend()
plt.grid()

plt.show()
````

## Nivel 4 — Ejemplo Aplicado: Cálculo de Notas

````python
# Simulación de notas de un curso
# Filas = estudiantes, Columnas = [Tarea1, Tarea2, Examen parcial, Examen final]
np.random.seed(42)  # Para reproducibilidad

notas = np.random.randint(40, 101, size=(8, 4))
pesos = np.array([0.15, 0.15, 0.30, 0.40])  # Pesos de cada evaluación

nombres = ["Ana", "Carlos", "Diana", "Eduardo", "Fátima", "Gabriel", "Helena", "Iván"]

print("Notas (Tarea1 | Tarea2 | Parcial | Final):")
print("-" * 50)
for nombre, fila in zip(nombres, notas):
    print(f"{nombre:<10} {fila}")

# Calcular nota final ponderada para cada estudiante
notas_finales = notas @ pesos
print(notas_finales)
````

````python
print("\n--- Resultados Finales ---")
for nombre, nota in zip(nombres, notas_finales):
    estado = "APROBADO" if nota >= 61 else "REPROBADO"
    print(f"{nombre:<10} Nota: {nota:.1f}  →  {estado}")

print(f"\nPromedio del curso: {np.mean(notas_finales):.2f}")
print(f"Nota más alta:      {np.max(notas_finales):.2f}")
print(f"Nota más baja:      {np.min(notas_finales):.2f}")
print(f"Aprobados: {np.sum(notas_finales >= 61)} / {len(notas_finales)}")
````
